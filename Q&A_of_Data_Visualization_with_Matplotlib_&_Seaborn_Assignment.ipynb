### **Netflix Movies and TV Shows:**

The dataset contains information about movies and TV shows available on Netflix as of 2021. Your task is to perform the following actions based on the data:

1. Identify missing values: Check if there are any missing values in the dataset and decide how to handle them. For example, if the "director" column has missing values, we can either delete those rows or fill in the missing values with "unknown".

2. Data Cleaning: Clean the dataset to remove any duplicates, inconsistencies or irrelevant columns. For example, we can remove the "show_id" and "description" columns as they may not be relevant for our analysis.

3. Data Exploration: Use Matplotlib to create visualizations that explore the data in meaningful ways. For example, we can create a bar chart to show the number of movies vs TV shows available on Netflix, or a line chart to show the trend in the number of new titles added each year.

4. Analyze Genres: Analyze the genres of movies and TV shows available on Netflix using bar charts or pie charts. For example, we can create a bar chart to show the top 10 most popular genres on Netflix.

5. Analyze Ratings: Analyze the ratings of movies and TV shows available on Netflix using histograms or box plots. For example, we can create a histogram to show the distribution of ratings on Netflix.

6. Insights and Conclusions: Based on the visualizations created, draw insights and conclusions about the dataset. For example, we may find that Netflix has a larger selection of TV shows than movies, or that the most popular genres on Netflix are drama and comedy.

netflix (1).png

```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('netflix_titles.csv')

# Identify missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
print()

# Data Cleaning
df_cleaned = df.drop(columns=['show_id', 'description']) 

# Data Exploration
# Bar chart: Number of movies vs TV shows
plt.figure(figsize=(6, 4))
df_cleaned['type'].value_counts().plot(kind='bar')
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
print()
# Line chart: Trend in the number of new titles added each year
df_cleaned['year_added'] = pd.to_datetime(df_cleaned['date_added']).dt.year
title_counts_per_year = df_cleaned['year_added'].value_counts().sort_index()
plt.figure(figsize=(6, 4))
plt.plot(title_counts_per_year.index, title_counts_per_year.values, marker='o')
plt.title('Number of New Titles Added each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(title_counts_per_year.index)
plt.grid(True)
plt.show()
print()
# Analyze Genres
genres = df_cleaned['listed_in'].str.split(', ')
genre_counts = {}
for genre_list in genres:
    for genre in genre_list:
        if genre in genre_counts:
            genre_counts[genre] += 1
        else:
            genre_counts[genre] = 1
top_genres = pd.Series(genre_counts).sort_values(ascending=False).head(10)

# Bar chart: Top 10 most popular genres
plt.figure(figsize=(6, 4))
top_genres.plot(kind='bar')
plt.title('Top 10 Most Popular Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
print()
# Analyze Ratings
plt.figure(figsize=(6, 4))
df_cleaned['rating'].value_counts().plot(kind='hist', bins=15)
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

### **US Accidents:**

For this activity, we will be using the "US Accidents" dataset. The dataset contains information about traffic accidents in the United States from 2016 to 2020.

üëâ Instructions:
1. Load the dataset into a pandas DataFrame and check for missing values.

2. Identify the top 10 states with the highest number of accidents and create a bar chart using matplotlib to visualize the data.

4. Identify the top 10 cities with the highest number of accidents and create a horizontal bar chart using matplotlib to visualize the data.

accident-insurance.png

```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('US_Accidents_Dec21_updated.csv')

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Identify the top 10 states with the highest number of accidents
top_10_states = df['State'].value_counts().head(10)

# Bar chart: Top 10 states with the highest number of accidents
plt.figure(figsize=(6, 4))
top_10_states.plot(kind='bar')
plt.title('Top 10 States with the Highest Number of Accidents')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Scatter plot: Location of accidents (longitude and latitude)
plt.figure(figsize=(6, 4))
plt.scatter(df['Start_Lng'], df['Start_Lat'], s=1, alpha=0.5)
plt.title('Accident Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Identify the top 10 cities with the highest number of accidents
top_10_cities = df['City'].value_counts().head(10)

# Horizontal bar chart: Top 10 cities with the highest number of accidents
plt.figure(figsize=(6, 4))
top_10_cities.sort_values().plot(kind='barh')
plt.title('Top 10 Cities with the Highest Number of Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('City')
plt.show()

# Pie chart: Distribution of accidents by severity
severity_counts = df['Severity'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Accidents by Severity')
plt.show()

# Line chart: Trend of accidents over time
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
accidents_per_year = df['Start_Time'].dt.year.value_counts().sort_index()
plt.figure(figsize=(6, 4))
plt.plot(accidents_per_year.index, accidents_per_year.values, marker='o')
plt.title('Trend of Accidents over Time')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(accidents_per_year.index)
plt.grid(True)
plt.show()

```

### **Data Visualization Task:**

Your task is to create a data visualization using Seaborn to explore the relationship between age, gender, and survival status of the passengers on the Titanic. Perform the following actions based on the data:

1. Identify missing values: Check for any missing values in the Age column.

2. Fill in missing values: If there are missing values in the Age column, use appropriate techniques such as imputation or interpolation to fill in the missing values.

3. Inconsistent formats: Check if there are any inconsistencies in the formatting of the gender column. For example, if the gender is encoded as "male" and "female", ensure that it is consistent throughout the dataset.

**Note:** You can use Seaborn functions such as countplot(), barplot(), boxplot(), scatterplot(), heatmap(), etc. to create the visualizations. Use appropriate Seaborn styling options such as color palettes, font sizes, and plot styles to make the visualizations visually appealing and informative.

cruise.png

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Check for missing values in the Age column
missing_age_count = df['age'].isnull().sum()
print("Missing values in Age column:", missing_age_count)

# Fill in missing values in the Age column using mean imputation
df['age'].fillna(df['age'].mean(), inplace=True)

# Check for inconsistencies in the formatting of the gender column
gender_counts = df['sex'].value_counts()
print("Gender counts:\n", gender_counts)

# Convert the data type of the Age column to float
df['age'] = df['age'].astype(float)

# Create a box plot to identify outliers in age
sns.boxplot(x=df['age'])
plt.xlabel('Age')
plt.title('Distribution of Passenger Age')
plt.show()

# Create a count plot to explore the relationship between gender and survival status
sns.countplot(x='sex', hue='survived', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survival Count by Gender')
plt.legend(['Did not survive', 'Survived'])
plt.show()

# Create a bar plot to explore the relationship between age, gender, and survival status
sns.barplot(x='sex', y='age', hue='survived', data=df)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Average Age by Gender and Survival Status')
plt.legend(['Did not survive', 'Survived'])
plt.show()

# Create a scatter plot to visualize the relationship between age, fare, and survival status
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare by Survival Status')
plt.legend(['Did not survive', 'Survived'])
plt.show()

# Create a heatmap to examine the correlation between age, gender, and survival status
correlation_matrix = df[['age', 'sex', 'survived']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix: Age, Gender, and Survival Status')
plt.show()
```

```python
# Write your code here
import seaborn as sns

# Load the Iris Species dataset
iris = sns.load_dataset('iris')

# Explore the data
print(iris.head())     # Display the first few rows of the dataset
print(iris.info())     # Get information about the dataset
print(iris.describe()) # Statistical summary of the dataset

# Data visualization
# Scatter plot: Sepal length vs. Sepal width
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

# Box plot: Petal length by species
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()

# Violin plot: Petal width by species
sns.violinplot(data=iris, x='species', y='petal_width')
plt.title('Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.show()

# Pair plot: Scatter plot matrix of all variables
sns.pairplot(data=iris, hue='species')
plt.suptitle('Pair Plot of Iris Variables')
plt.show()

# Heatmap: Correlation matrix of all variables
correlation = iris.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
